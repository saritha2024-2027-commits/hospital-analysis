```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("hospital data analysis.csv")
print(df.columns.tolist())
df.head()
df.info()
df.describe()
df.isnull().sum()
# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Fill missing values
df.fillna("Unknown", inplace=True)

bins = [0,18,35,50,100]
labels = ['Child','Young Adult','Adult','Senior']

df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

age_count = df['Age_Group'].value_counts()
print(age_count)
plt.figure(figsize=(6,4))
age_count.plot(kind='pie', autopct='%1.1f%%')
plt.title("Patients by Age Group")
plt.ylabel('')
plt.show()
readmission_count = df['Readmission'].value_counts()

print(readmission_count)
import matplotlib.pyplot as plt

readmission_count = df['Readmission'].value_counts()

plt.figure(figsize=(6,4))
readmission_count.plot(kind='bar')
plt.title('Patient Readmission Count')
plt.xlabel('Readmission')
plt.ylabel('Number of Patients')
plt.show()
plt.figure(figsize=(10,5))

```
