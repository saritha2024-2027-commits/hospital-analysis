```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("hospital data analysis.csv")
df.head()
df.info()
df.describe()
df.isnull().sum()
# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Fill missing values
df.fillna("Unknown", inplace=True)

# Check null values again
df.isnull().sum()
disease_count = df["disease"].value_counts()
print(df.columns.tolist())
print(disease_count)
plt.figure(figsize=(8,5))
disease_count.plot(kind='bar')
plt.title("Patient Count by Disease")
plt.xlabel("Disease")
plt.ylabel("Patients")
plt.show()
bins = [0,18,35,50,100]
labels = ['Child','Young Adult','Adult','Senior']

df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

age_count = df['Age_Group'].value_counts()
print(age_count)
plt.figure(figsize=(6,4))
age_count.plot(kind='pie', autopct='%1.1f%%')
plt.title("Patients by Age Group")
plt.ylabel('')
plt.show()
df['Admission_Date'] = pd.to_datetime(df['Admission_Date'])

monthly_admissions = df.groupby(
    df['Admission_Date'].dt.month
).size()

print(monthly_admissions)
plt.figure(figsize=(8,5))
monthly_admissions.plot(marker='o')
plt.title("Monthly Patient Admissions")
plt.xlabel("Month")
plt.ylabel("Admissions")
plt.grid(True)
plt.show()
daily_trend = df.groupby(
    df['Admission_Date'].dt.date
).size()

plt.figure(figsize=(10,5))
daily_trend.plot()
plt.title("Daily Patient Trend")
plt.xlabel("Date")
plt.ylabel("Patients")
plt.show()

```
